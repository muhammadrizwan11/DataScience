# Columns Opreations

```python
import pandas as pd
```

```python
data = {
    'City': ['Karachi', 'Lahore', 'Islamabad', 'Faisalabad', 'Rawalpindi', 'Multan', 'Quetta', 'Peshawar', 'Gujranwala', 'Sialkot'],
    'Population_Millions': [14.91, 11.13, 1.01, 3.2, 2.1, 1.87, 1.01, 1.97, 1.44, 0.89],
    'Area_SqKm': [591, 1772, 906, 1288, 259, 133, 2659, 1265, 319, 3.027],
    'GDP_Billion_USD': [120, 84, 58, 40, 40, 30, 14, 22, 21, 16],
    'Airports': [2, 1, 1, 1, 1, 1, 1, 1, 1, 1]
}
df = pd.DataFrame(data)
df
```

# Renaming Columns:
You can rename columns using the rename() method.

```python
df.rename(columns={'Population_Millions':'Peoples'},inplace=True)
```

```python
df
```

# Sorting by Column:
You can sort the DataFrame based on a specific column using the sort_values() method.

```python
df.sort_values(by='GDP_Billion_USD')
```

# Filtering Based on Column Values:
You can filter rows based on specific column values using boolean indexing.

```python
df[df['GDP_Billion_USD']<40]
```
