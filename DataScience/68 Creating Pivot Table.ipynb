# Pandas Pivot_table()
The pandas.pivot_table() function is a powerful tool in the Python pandas library for reshaping and summarizing data. It allows you to create pivot tables, which are a way of summarizing and aggregating data from a DataFrame to gain insights or perform analysis. Pivot tables are especially useful for summarizing data when you have multiple variables and want to see the relationships between them.

```python
import pandas as pd
```

```python
df = pd.read_excel('Filmes.xlsx')
```

```python
df
```

```python
df.pivot_table(index='Type')
```

```python
df.pivot_table(index='Type',columns='Year')
```

```python
df.pivot_table(index='Type',columns='Year',aggfunc='sum')
```

```python
df.pivot_table(index='Type',columns='Year',aggfunc='sum',fill_value=0)
```

```python
df.pivot_table(index='Type',columns='Year',aggfunc='sum',fill_value=0,margins=True)
```
