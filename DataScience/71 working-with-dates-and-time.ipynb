# Working with Dates

```python
date.info()
```

```python
# Converting to datetime datatype
date['date'] = pd.to_datetime(date['date'])
```

```python
date.info()
```

## 1. Extract year

```python
date['Year'] = date['date'].dt.year
date.head()
```

## 2. Extract Month

```python
date['month'] = date['date'].dt.month
date.head()
```

## Extract Month Name

```python
date['month_name'] = date['date'].dt.month_name()
date.head()
```

## Extract Days

```python
date['day'] = date['date'].dt.day
date.head()
```

```python
# day of week
date['day_of_week'] = date['date'].dt.dayofweek
date.head()
```

```python
# day of week - name
date['day_name'] = date['date'].dt.day_name()
date.head()
```

```python
# is weekend?
date['weekend'] = np.where(date['day_name'].isin(['Sunday','Saturday']),1,0)
date.head()
```

## Extract week of the year

```python
date['week_of_year'] = date['date'].dt.week
date.head()
```

## Extract Quarter

```python
date['quarter'] = date['date'].dt.quarter
date.head()
```

## Extract Semester

```python
date['semester'] = np.where(date['quarter'].isin([1,2]),1,2)
date.head()
```
