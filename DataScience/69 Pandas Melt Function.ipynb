# Melt() Function

The melt function in Pandas is a powerful tool for reshaping your data. It allows you to convert data from a wide format to a long format, making it easier to analyze and visualize.

```python
import pandas as pd
```

```python
df  = pd.read_excel('Filmes.xlsx')
```

```python
df
```

```python
help(pd.melt)
```

```python
df.melt()
```

```python
df.melt(id_vars = 'Budget(M$)')
```

```python
df.melt(id_vars = 'Budget(M$)',value_vars= 'Year')
```

```python
df.melt(id_vars = 'Budget(M$)',value_vars= 'Year',var_name='VR')
```

```python
df.melt(id_vars = 'Budget(M$)',value_vars= 'Year',var_name='VR',value_name='Year')
```

```python
df.melt(id_vars = 'Budget(M$)',value_vars= 'Year',var_name='VR')
```
