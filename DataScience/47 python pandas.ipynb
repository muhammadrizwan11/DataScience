# data structures in pandas

Series: A one-dimensional labeled array, capable of holding various data types (e.g., integers, strings, floats). It is similar to a NumPy array but with labeled indexes, allowing for more flexible data manipulation.

DataFrame: A two-dimensional labeled data structure, resembling a spreadsheet or SQL table, consisting of rows and columns. It is the most commonly used data structure in Pandas and allows for efficient data handling, cleaning, and analysis.
