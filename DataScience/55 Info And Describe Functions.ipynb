# Basic Exploration Functions:

# info()

The info() function in pandas provides a  summary of a DataFrame, including information about its data types, non-null values, memory usage, and more

```python
import pandas as pd
data = {
    'Student_ID': [101, 102, 103, 104, 105],
    'Age': [20, 21, 19, 22, 20],
    'Height_cm': [170, 175, 162, 180, 168],
    'Weight_kg': [65, 70, 55, 75, 62],
    'Math_Score': [85, 90, 75, 92, 88],
    'English_Score': [78, 85, 70, 88, 82]
}
df = pd.DataFrame(data)
df
```

```python
df.info()
```

# describe():

The describe() function generates summary statistics for numeric columns in the DataFrame.

```python
df.describe()
```
